```python
# Imports
import ipytest
import pandas as pd
import import_ipynb
import data_preprocessing

ipytest.autoconfig()
```

```python
%%ipytest

# def test_strip_punctuation():
#     input = 'lorem!"#$%&\'()*+, -./:;<=>?@[\]^_`{|}~ipsum'
#     expected = 'lorem ipsum'
#     assert data_preprocessing.strip_punctuation(input) == expected

def test_strip_non_alphanumeric():
    input = 'klasyfikacja!"(),012tekstu'
    expected = 'klasyfikacja        tekstu'
    assert data_preprocessing.strip_non_alphanumeric(input) == expected

def test_replace_whitespace():
    input = 'klasyfikacja        tekstu'
    expected = 'klasyfikacja tekstu'
    assert data_preprocessing.replace_whitespace(input) == expected

def test_delete_escape_chars():
    input = 'klasyfikacja\\ntekstu\nz\twykorzystaniem\ruczenia maszynowego'
    expected = 'klasyfikacja tekstu z wykorzystaniem uczenia maszynowego'
    assert data_preprocessing.delete_escape_chars(input) == expected

def test_delete_stop_words():
    input = 'przy klasyfikacji tekstu z wykorzystaniem przede wszystkim uczenia maszynowego'
    expected = ['klasyfikacji', 'tekstu', 'wykorzystaniem', 'uczenia', 'maszynowego']
    assert data_preprocessing.delete_stop_words(str.split(input)) == expected

def test_lemmatise():
    input = 'klasyfikacji tekstu wykorzystaniem uczenia maszynowego'
    df_input = pd.DataFrame([[str.split(input)]], columns=['Text'])
    expected = 'klasyfikacja tekst wykorzystać uczyć maszynowy'
    df_expected = pd.DataFrame([[expected]], columns=['Text'])
    assert df_expected.equals(data_preprocessing.lemmatise(df_input))
```
